# Exploring the Effect of Rising Minimum Wage in San Francisco

#### Background

* Since July 24, 2009, the federal government has mandated a nationwide minimum wage of 7.25 per hour.
* In 2004 the City of San Francisco passed an ordinance linking the minimum wage to inflation (consumer price index).
* On November 4, 2014, San Francisco voters passed Proposition J, raising the minimum wage to 15.00 by 2018.

As municipalities across the country debate whether and how much to raise MW we wanted to explore what, if any, impact we could see in SF. We took a look at the Wage and Employment relationship in 3 industry groups that employ a larger share of MW workers. We also wanted to see if we could find any impact on community in regard to crime. We looked at the 4 largets crime groups for a broad overview on crime trends.
